```python
import pandas as pd
```

```python
# A pandas series is like a column in table. 
# It is a 1-d array which can hold any type of data

s=[1,2,3,4]
my_series=pd.Series(s)
print(my_series)

print(my_series[0],my_series[1])
```

```python
# Create Labels 
my_series1=pd.Series(['Uday','Nikhil','Abhay'],index=['a','b','c'])
print(my_series1)

print(my_series1['a'],my_series1['b'],my_series1['c'])
```
